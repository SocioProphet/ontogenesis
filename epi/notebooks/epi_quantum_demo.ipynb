# Epi‑Onto‑Learning — Quantum Lane Demo (Reference)

This notebook is a **reference** in the ontology repo.

Goals:
1. Estimate an **SU(2) charge proxy** using an Estimator (Aer by default).
2. Run a **64‑node heavy‑hex‑shaped QAOA toy** (stub) to pick a routing schedule.
3. Publish a **PDF report** + JSON provenance sidecar.

> Note: Hardware is opt‑in and must be fully logged in the report.


```python
from epi.tools.quantum_adapter import QuantumChargeProjector
from epi.tools.report_publish import Provenance, write_report
from datetime import datetime
from pathlib import Path

projector = QuantumChargeProjector(backend='aer', provider='qiskit', mitigation=('m3','zne'))

# Stub: in a training repo, 'circuit' would be a real qiskit.QuantumCircuit
circuit = None
res = projector.estimate(pauli_op='Z0', circuit=circuit)
res
```

```python
# QAOA toy placeholder
def qaoa_pick_schedule(n=64, p_layers=2):
    # In a production integration, encode an Ising/MaxCut routing objective.
    return {'picked_schedule':'stub', 'n':n, 'p_layers':p_layers}

schedule = qaoa_pick_schedule()
schedule
```

```python
out_pdf = Path('epi_artifacts/epi_quantum_demo_report.pdf')
prov = Provenance(
    commit='(set-by-ci)',
    signer='(set-by-ci)',
    backend=res.backend,
    provider=res.provider,
    mitigation=list(res.mitigation),
    created_utc=datetime.utcnow().isoformat()+'Z',
)

body = [
    f"SU(2) charge proxy estimate: mean={res.mean} var={res.variance}",
    f"QAOA schedule pick: {schedule}",
]

write_report(out_pdf, prov, title='Epi‑Onto‑Learning Quantum Lane Demo', body_lines=body)
out_pdf
```
